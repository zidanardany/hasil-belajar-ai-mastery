**Image Cropping**

```python
'''
dokumentasi
https://docs.opencv.org/3.4/da/d54/group__imgproc__transform.html
'''

patch = (150, 150)
center = (100, 90)

cropped_image = cv2.getRectSubPix(circles, patch, center)

cv2_imshow(cropped_image)
```

```python
def crop(img, center, width, height):
  return cv2.getRectSubPix(img, (width, height), center)

cropped_image = crop(circles, center=(110, 80), width=150, height=150)
cv2_imshow(cropped_image)
```

**Rotation**

```python
# shape yang dibaca adalah height dulu, baru width
circles.shape
```

```python
# membuat center dengan program
center = int(circles.shape[1]/2), int(circles.shape[0]/2)
center
```

```python
# image rotation
angle = 15
scale = 1.0

# getRotationMatrix2D() membaca width dulu, baru height
R = cv2.getRotationMatrix2D(center, angle, scale)
print(R)
```

```python
rotated_cat = cv2.warpAffine(circles, R, (center[0]*2, center[1]*2))

cv2_imshow(rotated_cat)
```
