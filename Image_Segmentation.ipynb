```python
def plot_img(images, titles):
  fig, axis = plt.subplots(nrows = 1, ncols = len(images), figsize=(15,15))
  for i, p in enumerate(images):
    axis[i].imshow(p, 'gray')
    axis[i].imshow(titles[i])
  plt.show()
```

**Adaptive Thresholding** 

[Image Thresholding](https://docs.opencv.org/4.x/d7/d4d/tutorial_py_thresholding.html)

```python
img = cv2.imread('corporate-espionage.jpg', 0)
ret, img_binary = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)

plt.imshow(img, 'gray')
plt.show()
plt.imshow(img_binary, 'gray')
plt.show()
```

```python
img_thresh_adaptive = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
plt.imshow(img_thresh_adaptive, 'gray')
plt.show()
```

```python
img_thresh_adaptive = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
plt.imshow(img_thresh_adaptive, 'gray')
plt.show()
```

**Otsu binarization** tools yang bisa digunakan untuk memodifikasi adaptive seperti yg digunakan gaussian c/mean c

```python
img = cv2.imread('Image_processing_pre_otsus_algorithm.jpg', 0)

plt.imshow(img, 'gray')
```

```python
# Otsu thresholding
ret, img_otsu = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

plt.imshow(img_otsu, 'gray')
```

```python
# Otsu thresholding
ret, img_otsu = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

plt.imshow(img_otsu, 'gray')
```

Basically, ret is a boolean regarding whether or not there was a return at all, at the frame is each frame that is returned. If there is no frame, you wont get an error, you will get None.
